```python
def convert_time(time):
    if 'am' in time:
        if len(re.findall('(\d+).(\d+)am', time)) > 0:
            n_1,n_2 = re.findall('(\d+).(\d+)am', time)[0]
            if int(n_1)%12 < 10:
                return '0' + str(int(n_1)%12) + n_2
            else:
                return n_1+n_2
        else:
            n = re.findall('(\d+)am', time)[0]
            if int(n)%12 < 10:
                return '0' + str(int(n)%12) + '00'
            else:
                return n + '00'
    elif 'pm' in time:
        if len(re.findall('(\d+).(\d+)pm', time)) > 0:
            n_1,n_2 = re.findall('(\d+).(\d+)pm', time)[0]
            return str(int(n_1)%12+ 12) + n_2
        else:
            n = re.findall('(\d+)pm', time)[0]
            return str(int(n)%12+ 12) + '00'
    else:
        return None
```
